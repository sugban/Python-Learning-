### Seaborn
* It can be considered as an extension of Matplotlib as it is built on top of that,so that means Seaborn is complimentary to Matplotlib. 
* It specifically targets statistical data visualization.
* Easy plotting 
* To import :
<pre>
import seaborn as sns

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading csv file and creating a dataframe

df = pd.read_csv('updated_nba.csv')
df.head()
```

```python
df.info()
```

```python
height = pd.DataFrame(df['Height'].value_counts()).reset_index()
height.columns=['Height','Total participants']
height
```

### Seaborn with matplotlib

```python
# Line plot with matplotlib

plt.plot(height['Height'],height['Total participants'])
plt.xlabel("Height")
plt.ylabel("Total number of participants")
plt.title("Number of participants of different height range")

plt.show()
```

Seaborn supports various themes that can make styling the plots really easy 

By using the **set_style()** function of Seaborn we can set any of the themes available on Seaborn library (Darkgrid, Whitegrid, Dark, White, and Ticks)

```python
# we can style this plot using the set_style() function

plt.plot(height['Height'],height['Total participants'])

sns.set_style("dark")

plt.show()
```

If you need your plots for talks, posters, on paper or in notebooks, you might want to have larger or smaller plots. Seaborn allows us to control the plot elements by using the **set_context()** function

```python
# we can set context to "paper"

plt.plot(height['Height'],height['Total participants'])
sns.set_style("whitegrid")

sns.set_context("paper")

plt.show()
```

### Seaborn plots

```python
# line chart
sns.lineplot(data=height, x='Height', y='Total participants')
```

Seaborn is a library that uses Matplotlib underneath to plot graphs.

```python
# to give it a title
sns.lineplot(data=height, x='Height', y='Total participants')

# setting the title using Matplotlib
plt.title('Total participants')
  
plt.show()
```

```python
# setting the limit of x axis of the plot

sns.lineplot(data=height, x='Height', y='Total participants')
plt.title('Total participants')

plt.xlim(5)       # it will start taking data after 5th row

plt.show()
```

**Count plots** take in a categorical list and return bars that represent the number of list entries per category. 

```python
sns.countplot(x=df['Height'])
```

```python
df1=pd.read_csv('data.csv')
df1.head()
```

**hue** allows us to easily make subgroups within Seaborn plots


---

**relplot()** stands for relational plot and enables you to visualize the relationship between two quantitative variables using either scatter plots or line plots. 

```python
sns.relplot(data=df1,x='Calories', y='Maxpulse', hue="Duration")
sns.set_style('white')
```

 Using "relplot()" gives us the ability to create subplots in a single figure.

```python
# # Change to make subplots based on study time
sns.relplot(data=df, x='Age', y='Salary', hue='Team', col='Position')
```

```python
# Change this scatter plot to arrange the plots in rows
sns.relplot(data=df, x='Age', y='Salary', row='Position')

# Show plot
plt.show()
```

```python
sns.relplot(data=df, x='Age', y='Salary', hue='Team', col='Position',kind='line')
```

```python
# sns.histplot() makes a histogram

sns.histplot(df1['Calories'])
```

**Distplot** is also a distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.


```python
sns.distplot(df1['Calories'].value_counts())
```

```python
# Use sns.boxplot() to make boxplots
sns.boxplot(df['Position'],df['Salary'])
```
