### Matplotlib
* Matplotlib is a graph plotting library that serves as a visualization utility
* Matplotlib consists of several plots like line, bar, scatter, histogram etc.
* Most of the Matplotlib utilities lies under the pyplot submodule,that's why we mostly import as plt alias
* We can import matplotlib with pyplot in two ways:
<pre>
from matplotlib import pyplot as plt
or
import matplotlib.pyplot as plt 
</pre>

* Visualization helps to understand trends, patterns, and to make correlations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading csv file and creating a dataframe

df = pd.read_csv('Files/nba.csv')
df
```

```python
df.info()
```

```python
# how to handle missing value in Pandas

df.dropna(inplace=True)  # it will drop all the rows in which any column has missing value
df.info()
```

- **plot() function** is used to draw points (markers) in a diagram.By default, it draws a line from point to point.
- It takes parameters for specifying points; first parameter is an array containing the points on the x-axis, and the second parameter is an array containing the points on the y-axis

```python
# extracting particular data from height column

height = df['Height'].value_counts()
print(height)
```

```python
# draw two points in the diagram

# x-axis values
x = height.loc[['5-9','5-11']]

# y-axis values
y = np.array(x.index)

# generating plot with two point position
plt.plot(x,y,'o') # 'o' is a marker, to emphasize each point

# to display the diagram
plt.show()
```

#### Labelling plot

We can plot even many points with having the same number of points in both axis, so if x-axis has 4 elements, y-axis also should have 4 elements

```python
# Line plot

# x-axis values
x = np.array(height.index)

# y-axis values
y = height.values

# generating line plot with plot() method
plt.plot(x,y)

# Labeling x and y - axis

plt.xlabel("Height")
plt.ylabel("Total number of participants")

# Title of the plot

plt.title("Number of participants of different height range")

plt.show()
```

### Changing size of the plot

- We can optimize plot's size as well
- There are various ways to change the default plot size as per our required dimensions or resize a given plot


---
1. With set_figheight() and set_figwidth() methods

They are used to adjust the width and height of the plot by passing a size number as argument

```python
# plotting a line plot after changing it's width and height
fig = plt.figure()

fig.set_figwidth(15)
fig.set_figheight(5)

print("Plot after re-sizing: ")
plt.plot(x, y)
plt.show()
```

 2. With figsize() method

It takes two parameters- width and height (in inches)

```python
# changing the size of figure to 2X2
plt.figure(figsize=(10, 5))

plt.show(plt.plot(x, y))
```

3. With the default rcParams

The default size of a figure can be permanently changed as per our needs by setting the figure.figsize

```python
# changing the rc parameters and plotting a line plot
plt.rcParams['figure.figsize'] = [10, 5]
  
plt.show(plt.plot(x, y))
```

```python
# Bar plot

# bar() method to generate bar plot
plt.bar(x,y)
  
# function to show the plot
plt.show()
```

3. Scatter Plot

It is a classic and fundamental plot to study the relationship between two variables

If you have multiple groups in your data, you can visualise each group in different color

```python
# Scatter plot with scatter() method

plt.scatter(x, y)
  
# function to show the plot
plt.show()
```

```python
df1=pd.read_csv('Files/data.csv')
df1
```

```python
# If you want to plot multiple lines in one chart, you can call the plot() function multiple times

plt.scatter(df1['Calories'],df1['Pulse'],marker='o')
plt.scatter(df1['Calories'],df1['Maxpulse'],marker='+')

plt.show()
```

### Adding Legend

```python
# legend() method helps adding labels to two different plots

plt.scatter(df1['Calories'],df1['Pulse'],marker='o')
plt.scatter(df1['Calories'],df1['Maxpulse'],marker='+')

# legend()
plt.legend(["Pulse", "Maxpulse"], loc ="lower right")

plt.show()
```

```python
# legend with labelling, while generating plot

plt.scatter(df1['Calories'],df1['Pulse'],label='Pulse')
plt.scatter(df1['Calories'],df1['Maxpulse'],label='Maxpulse')

# Function add a legend  
plt.legend()
  
# function to show the plot
plt.show()
```

```python
# pie chart

y = df['Height'].value_counts()

plt.pie(y) # pie chart takes only numeric value 
plt.show() 
```

```python
# pie chart with label

y = df['Height'].value_counts()

plt.pie(y,labels=y.index) 
plt.show() 
```

```python
# pie chart with legend

y = df['Height'].value_counts()

plt.pie(y)

plt.legend(labels=y.index,loc='right')
plt.show() 
```

### Histogram

### Subplots

- you can draw multiple plots in one figure with subplot() method
- It takes three arguments that describes the layout of the figure; row,column and the index number of the current plot
<pre>
plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot
plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot
</pre>

```python
# plot 1
plt.subplot(1,2,1)
plt.scatter(df1['Calories'],df1['Pulse'])

# plot 2
plt.subplot(1,2,2)
plt.scatter(df1['Calories'],df1['Maxpulse'])

  
# function to show the plot
plt.show()
```

<pre>
plt.subplot(2, 1, 1)
#the figure has 2 row, 1 column, and this plot is the first plot
plt.subplot(2, 1, 2)
#the figure has 2 row, 1 column, and this plot is the second plot
</pre>

```python
# plot 1
plt.subplot(2,1,1)
plt.scatter(df1['Calories'],df1['Pulse'])

# plot 2
plt.subplot(2,1,2)
plt.scatter(df1['Calories'],df1['Maxpulse'])

  
# function to show the plot
plt.show()
```
