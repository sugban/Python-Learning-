### Excel file - read, create, update 

* xlrd module is used to extract data from a spreadsheet
<pre>
import xlrd
</pre>


```python
# importing module

import xlrd
```

```python
# Loading Excel file

exl_file = xlrd.open_workbook('Newcustomer.xlsx')
```

*There could be multiple workbooks(sheets) inside a single excel file. So to work on sheet by sheet or one sheet, we will use **'sheet_by_index()'** function and pass the index number of the sheet we want to work on*

```python
# Storing the first sheet into a variable

purchase = exl_file.sheet_by_index(0)
```

```python
# Printing various cell values

print("Value of 0-0 cell: ",purchase.cell_value(0, 0)) # printing value from first row and first column

print("Value of 20-1 cell: ",purchase.cell_value(20, 0))  # printing value from 20th row and first column
```

```python
# Get the no of rows and columns

print("Number of Rows: ", purchase.nrows)
print("Number of Columns: ",purchase.ncols)
```

```python
# Get all column names

print("Columns in the worksheet are: ")

for col in range(purchase.ncols):
    print(purchase.cell_value(0,col))
```

```python
# to get data from a particular column - the first column

purchase.cell_value(0, 0)
 
for row in range(purchase.nrows):
    print(purchase.cell_value(row, 0))
```

```python
# Get first 10 rows for 3 columns

for row in range(11): # looping 10 times to get the data from first 10 rows
    for col in range(3): # looping 3 times to get the data in first 3 columns

        print(purchase.cell_value(row,col), end="\t ")
    print()
```

```python
# to extract a particular row value - second row (without header first row)
 
print(purchase.row_values(1))
```

### Writing to an excel sheet

* to write and modify data in an excel sheet, we have to import a module xlwt
<pre>
import xlwt
</pre>

```python
# importing module

import xlwt
from xlwt import Workbook
```

```python
# creating a new workbook

wb = Workbook()
```

```python
# to create sheet, we'll use add_sheet() method

sheet1 = wb.add_sheet('Sheet 1')
```

```python
# adding rows and columns in sheet 1

sheet1.write(0, 0, 'Student name')
sheet1.write(0, 1, 'DOB')
sheet1.write(0, 2, 'Course')
sheet1.write(0, 3, 'Year')
sheet1.write(1, 0, 'Allan')
sheet1.write(1, 1, '20-01-1999')
sheet1.write(1, 2, 'Data Science')
sheet1.write(1, 3, '2020')
sheet1.write(2, 0, 'RAJ')
sheet1.write(2, 1, '12-05-1996')
sheet1.write(2, 2, 'UX')
sheet1.write(2, 3, '2018')

# giving workbook a name and saving it 

wb.save('Student.xlsx')
```

```python
# to open to read

exl_file = xlrd.open_workbook('Student.xlsx')

student = exl_file.sheet_by_index(0)
```

```python
print(student.nrows)
print(student.ncols)
```

```python
# to read the file all content

for row in range(student.nrows):
  print(student.row_values(row))
```
