### Loops

* Loops are the constructs that repeatedly execute a piece of code. 
* Loops are used in many situations like going through every element of a list, doing an operation on a range of values, etc.
* Basically loops is the repeatation of the same thing for multiple times.
-------------------------------------------------------------------------------------------------------------------------------
**There are two types of loops in Python: While and For**

### While Loop

**Syntax**

<pre>
while expression:
    statement(s)
</pre>

**while** loop is used to execute a block of statements repeatedly until a given a condition is satisfied. And when the condition becomes false, the statement(s) immediately is executed.

**Flow chart**

```python
i = 1
while i < 5:  # print i as long as i gets the values less than 5
    print(i)
    i += 1  # increasing i value by assigning operator (i=i+1)
```

### break Statement

* To break the loop we use break statement 
* With the break statement we can stop the loop, even if the while condition is true.

```python
# IF condition with break

i = 1
while i < 10:
    print(i)
    
    # exit the loop as i gets value 5
    if i == 5:  
        break   # break
    i += 1
```

### continue Statement

* To stop the current iteration, and continue with the next
* With the continue statement we can skip the current iterartion or loop as the condition becomes TRUE and it will start the loop again with next iteration

```python
# IF condition with continue

i = 0
while i < 10:
    i += 1
    
    # skip the loop as i gets value 5
    if i == 5:  
        continue   # skip and would not print as value becomes 5
    
    print(i)
```

### else Statement

* while loop executes the block until a condition is satisfied. When the condition becomes false, the statement immediately after the loop is executed. The else statement is only executed, when your while condition becomes false. 
* With the else statement we can still run a block of code, when the condition no longer is true

```python
i = 1
while i < 10:
    print(i)
    i += 1
    
# Print the message, as the condition of while loop gets FALSE
else:
    print("i gets the value 10 and the condition becomes FALSE.")
```

### Infinite Loop

* when you set the condition of while loop TRUE, it will run non stop (infinite)
* To stop infinite loop from running press the STOP button

```python
while True:
    print('Infinte Loop!')
```

### For Loop

**Syntax**
<pre>
for iter_var in sequence:
    statement(s)
</pre>

* for loop is used for iterating over a sequence, that can be a list, a tuple, a dictionary, a set, or a string.
* The for loop does not require an indexing variable to set beforehand, like while loop.

**Flow chart**

<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fresize.yandex.net%2Fmailservice%3Furl%3Dhttps%3A%252F%252Fi1.faceprep.in%252FCompanies-1%252Fflowchart-of-for-loop-in-python.png%26proxy%3Dyes%26key%3D751f30b03fa9aefecab85d737dd34db1&f=1&nofb=1" alt="image" width="400"/>


```python
name = 'Sugandh Bansal'

for i in name:
    print(i)
```

```python
# to print all the iterated value in same line

name = 'Sugandh Bansal'

for i in name:
    print(i,end=' ') # end='': to keep the pointer on the same line.
```

### range() Function with for Loop

* For loops can iterate over a sequence of numbers using the "range" function
* The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at -1 from the given number, i.e. range(10) and it will print 10 values starting from 0 to 9.

```python
for i in range(10):
    print(' Hello Python')
```

```python
n = 4
for i in range(0, n):
    print(i)
```

### break Statement
* With the break statement we can stop the loop before it has looped through all the items
* It brings control out of the loop

```python
for i in range(10):
        
    # Condition to break the loop
    if i == 5:
        break
        
    print(i,': Hello Python')
```

### continue Statement

* With the continue statement we can stop the current iteration of the loop, and continue with the next
* It will skip the current iterarted value and move to the next

```python
for i in range(10):
        
    # Condition to skip the loop, as it gets the value 5
    if i == 5:
        continue
        
    print(i,': Hello Python')
```

### else Statement

* We can also combine else statement with for loop like in while loop. 
* The differnece is that there is no condition in for loop based on which the execution will terminate, so the else statement will be executed immediately after for loop finishes execution


```python
for x in range(10):
    print(x)
else:
    print("For loop is completed.")
```

### pass statement

*for loops cannot be empty.
* If for some reason you have a for loop with no content, put in the pass statement to avoid getting an error.

```python
for i in 'Sugandh':
    pass
```
