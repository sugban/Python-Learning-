#### max_rows
* The number of rows returned is defined in Pandas option settings
* We can check system's maximum rows with **pd.options.display.max_rows** statement


```python
print(pd.options.display.max_rows) 
```

* the number is 60, which means that the DataFrame contains more than 60 rows
* Thus, whenever we print dataframe, it will by default show only first and last 5 rows
* We can increase the maximum number of rows to display the entire DataFrame by assigning a number to display option statement


```python
pd.options.display.max_rows = 999
```

#### Location based indexing (loc/iloc)

* DataFrame.loc[] method is used to retrieve rows from Pandas DataFrame
* Rows can also be selected by passing integer location to an iloc[] function

```python
# will get third row details
data.loc[3]
```

```python
# to get data details from second to forth row
data.loc[1:4]
```

```python
# To change the index
data.set_index('Name',inplace=True)
data
```

```python
# to get data details from selected row 
data.loc['Markel Brown']
```

```python
# to get a particular row and column
data.loc['Markel Brown','Position']
```

```python
# to get multiple rows
data.loc[['Markel Brown','Henry Sims']]
```

```python
# to get multiple rows and columns
data.loc[['Markel Brown','Henry Sims'],['Team','Position']]
```

**df.iloc** indexer is very similar to df.loc but only uses integer locations to make its selections

```python
data.iloc[1]
```

```python
data.iloc[4:8]
```

```python
# index can be reset by df.reset_index() method

data.reset_index(inplace=True)
data
```

```python
# to get selected rows

data.iloc[[10,15,20,25]]
```

```python
# to get selected row and column

data.iloc[[10,15,20,25],5:10]
```

```python
data.iloc[[10,15,20,25],[0,1,3,4,7,8]]
```
